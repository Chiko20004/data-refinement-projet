# 02 – Nettoyage des données




```python
import pandas as pd

df = pd.read_csv("../data/raw/dirty_cafe_sales.csv")
df.head()
```

```python
df.duplicated().sum()
```

Une vérification des doublons est nécessaire pour vérifier la qualité des données

```python
df.isnull().sum()
```

Plusieurs colonnes présentent des valeurs manquantes.
un traitement est donc nécessaire afin de préserver la
cohérence du dataset.


```python
df = df.dropna(subset=[
    "Item",
    "Quantity",
    "Price Per Unit",
    "Transaction Date"
])
```

```python
df["Payment Method"] = df["Payment Method"].fillna("Unknown")
df["Location"] = df["Location"].fillna("Unknown")
```

Les colonnes essentielles pour interpréter une transaction
(produit, quantité, prix et date) ont été traitées en priorité.
Les lignes contenant des valeurs manquantes sur ces champs ont
été supprimées, car elles ne permettent pas une lecture fiable
de la transaction.

À l’inverse, les colonnes secondaires comme la méthode de paiement
ou la localisation restent exploitables même en cas d’information
manquante. Les valeurs absentes ont donc été remplacées par
"Unknown" afin de conserver un maximum de données.


```python
df.isnull().sum()
```

Après le nettoyage, le jeu de données ne contient plus de valeurs
manquantes bloquantes. Les données sont désormais plus cohérentes
et prêtes pour la phase de transformation.


Conclusion

Cette étape de nettoyage a permis de corriger les principaux problèmes
de qualité identifiés lors de l’exploration, les valeurs manquantes. Le jeu de données obtenu est plus fiable et constitue une base saine pour les transformations à venir.
