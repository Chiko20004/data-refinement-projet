### Structure et types des données

Le jeu de données contient 10 000 lignes et 8 colonnes.  
Toutes les colonnes sont actuellement de type `object`, y compris celles
qui devraient représenter des valeurs numériques ou des dates, ce qui
indique un problème de typage des données.

Plusieurs colonnes présentent également des valeurs manquantes,
notamment pour les informations liées aux produits, aux paiements
et aux localisations.  
Ces observations confirment la nécessité d’une phase de nettoyage
et de transformation avant toute exploitation du jeu de données.


```python
df.columns
```

Description des colonnes

Le jeu de données contient les colonnes suivantes :

- Transaction ID : identifiant unique de la transaction  
- Item : produit vendu  
- Quantity : quantité vendue  
- Price Per Unit : prix unitaire du produit  
- Total Spent : montant total de la transaction  
- Payment Method : méthode de paiement utilisée  
- Location : lieu de la vente  
- Transaction Date : date de la transaction
